```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

colum=['id','Date','number of bedrooms','Area of the house(excluding basement)']
df=pd.read_csv('House Price India.csv',usecols=colum)
df.head
```

```python
df.isnull().sum()
```

# univariate analysis

```python
df['Area of the house(excluding basement)'].describe()
```

```python
plt.pie(df['number of bedrooms'].value_counts())
plt.show()
```

```python
sns.histplot(df['Area of the house(excluding basement)'], kde=True)
plt.title('Histogram of Area of the house(excluding basement)')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='Area of the house(excluding basement)', data=df)
plt.title('Box Plot of Area of the house(excluding basement)')
plt.xlabel('Area of the house(excluding basement)')
plt.show()
```

```python
sns.countplot(x='Area of the house(excluding basement)', data=df)
plt.title('Count Plot of Area of the house(excluding basement)')
plt.xlabel('Area of the house(excluding basement)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Bivariate

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
sns.scatterplot(x='number of bedrooms', y='Area of the house(excluding basement)', data=df)
```

```python
correlation = df['number of bedrooms'].corr(df['Area of the house(excluding basement)'])
print(f"Correlation between number of bedrooms and Area of the house(excluding basement) : {correlation}")
```

```python
plt.figure(figsize=(8, 6))  # Set the figure size (optional)
sns.boxplot(x='number of bedrooms', y='Area of the house(excluding basement)', data=df)
```

# Multivariate

```python
selected_columns = ['Date','number of bedrooms','Area of the house(excluding basement)']
sns.pairplot(df[selected_columns])
plt.show()
```

```python
correlation_matrix = df[selected_columns].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))  # Set the figure size (optional)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
